```python
import pandas as pd
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv")
df.to_sql("SPACEXTBL", con, if_exists='replace', index=False,method="multi")
```

```python
#DROP THE TABLE IF EXISTS

%sql DROP TABLE IF EXISTS SPACEXTABLE;
```

```python
%sql create table SPACEXTABLE as select * from SPACEXTBL where Date is not null
```

```python
%sql SELECT * FROM SPACEXTABLE
```

```python
%sql SELECT DISTINCT Launch_Site FROM SPACEXTABLE
```

```python
%sql SELECT * FROM SPACEXTABLE WHERE Launch_Site like "CCA%" LIMIT 5
```

```python
%sql SELECT SUM(PAYLOAD_MASS__KG_) FROM SPACEXTABLE WHERE Customer like "NASA (CRS)%"
```

```python
%sql SELECT AVG(PAYLOAD_MASS__KG_) FROM SPACEXTABLE WHERE Booster_Version LIKE "F9 v1.1%"
```

```python
%sql SELECT Date FROM SPACEXTABLE WHERE Landing_Outcome like "Success%" LIMIT 1
```

```python
%sql SELECT Booster_Version	FROM SPACEXTABLE WHERE 4000 < PAYLOAD_MASS__KG_ < 6000 AND  Landing_Outcome like "%Success (drone ship)%"
```

```python
%sql SELECT COUNT(Mission_Outcome) FROM SPACEXTABLE WHERE Mission_Outcome like "%Success%"
```

```python
%sql SELECT COUNT(Mission_Outcome) FROM SPACEXTABLE WHERE Mission_Outcome like "%Failure%"
```

```python
%sql SELECT Booster_Version FROM SPACEXTABLE  WHERE PAYLOAD_MASS__KG_ = (SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTABLE);
```

```python
%%sql SELECT 
    SUBSTR(Date, 6, 2) AS Month,
    Booster_Version,
    Launch_Site,
    Landing_Outcome 
FROM SPACEXTABLE 
WHERE 
    Landing_Outcome LIKE '%Failure (drone ship)%'
    AND SUBSTR(Date, 1, 4) = '2015';
```

```python
%%sql SELECT 
    Landing_Outcome, 
    COUNT(Landing_Outcome) AS Outcome_Count 
FROM SPACEXTABLE 
WHERE 
    Date BETWEEN '2010-06-04' AND '2017-03-20' 
GROUP BY Landing_Outcome 
ORDER BY Outcome_Count DESC;
```
